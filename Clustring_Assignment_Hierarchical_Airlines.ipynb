### Using Normalize Function


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
#import Dataset
df=pd.read_excel('C:/Users/Akash Kundu/Desktop/ExcelR/Assingments/EastWestAirlines.xlsx')
df
```

```python
df.info()
```

```python
df1=df.drop(['ID#'],axis=1)
df1
```

```python
# Normalize heterogenous numerical data 
df1_norm=pd.DataFrame(normalize(df1),columns=df1.columns)
df1_norm
```

```python
# Create Dendrograms
plt.figure(figsize=(15,8))
dendogram=sch.dendrogram(sch.linkage(df1_norm,'complete'))
```

```python
# Create Clusters (y)
hcluster=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hcluster
```

```python
y=pd.DataFrame(hcluster.fit_predict(df1_norm),columns=['clusterid'])
y['clusterid'].value_counts()
```

```python
# Adding clusters to dataset
df1['clusterid']=hcluster.labels_
df1
```

```python
df1.groupby('clusterid').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(15,8))
plt.scatter(df1['clusterid'],df1['Balance'],c=hcluster.labels_)
plt.show()
```
